```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count # 혀재 값을 반환하고, 호출될 때까지 중단
        count += 1

# 제너레이터 호출
counter = count_up_to(5)
for num in counter:
    print(num)
```

```python
def infinite_sequence():
    num = 0
    while True:
        yield num
        num += 1

gen = infinite_sequence()
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def  read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line
    
# 재내레이터 사용
for line in read_large_file('large_text_file.txt'):
    print(line, end='')
```

```python
# 리스트 컴프리헨션
squares = [x**2 for x in range(10)]
print(squares)

# 제너레이터 표현식
squares_gen = (x**2 for x in range(10))
print(next(squares_gen))
print(next(squares_gen))
```

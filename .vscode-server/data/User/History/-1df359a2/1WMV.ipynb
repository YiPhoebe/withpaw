```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count # 혀재 값을 반환하고, 호출될 때까지 중단
        count += 1

# 제너레이터 호출
counter = count_up_to(5)
for num in counter:
    print(num)
```

```python
def infinite_sequence():
    num = 0
    while True:
        yield num
        num += 1

gen = infinite_sequence()
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def  read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line
    
# 재내레이터 사용
for line in read_large_file('large_text_file.txt'):
    print(line, end='')
```

```python
from datetime import timedelta

# 현재 날짜로부터 5일 후의 날짜 계산
future_date = now + timedelta(days=5)
print("5일 후의 날짜:", future_date)

# 날짜와 시간 포맷팅
formatted_date = now.strftime("%Y-%m-%d %H:%M:%S")
print("포맷팅된 날짜와 시간:", formatted_date)

# 문자열을 날짜와 시간 객체로 변환
date_str = "2023-10-10 14:30:00"
parsed_date = datetime.strptime(date_str, "%Y-%m-%d %H:%M:%S")
print("변환된 날짜와 시간 객체:", parsed_date)
```

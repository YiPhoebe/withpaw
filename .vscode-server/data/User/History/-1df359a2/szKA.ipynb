```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count  # 현재 값을 반환하고, 호출될 때까지 중단
        count += 1

# 제너레이터 호출
counter = count_up_to(5)
for num in counter:
    print(num)
```

```python
# 캐릭터 생성
class Character:
    def __init__(self, name, health, attack, defense):
        self.name = name
        self.health = health
        self.attack = attack
        self.defense = defense
        self.exp = 0
        self.level = 1

    def take_damage(self, damage):
        self.health -= damage
        if self.health < 0:
            self.health = 0

    def is_alive(self):
        return self.health > 0
    
    def level_up(self):
        self.level += 1
        self.exp = 0
        self.health += 50
        self.attack += 5
        print(f"축하합니다! {self.name}이(가) 레벨 {self.level}로 올라갔습니다!")
        print(f"새로운 체력: {self.health}, 공격력: {self.attack}")
    
    

# 몬스터 클래스
class Monster:
    def __init__(self, name, health, attack):
        self.name = name
        self.health = health
        self.attack = attack

    def take_damage(self, damage):
        self.health -= damage
        if self.health < 0:
            self.health = 0
        
    def is_alive(self):
        return self.health > 0

# 캐릭터 생성
player = Character("유정", 100, 20, 10)
print(f"{player.name} 캐릭터가 생성되었습니다! 체력: {player.health}, 공격력: {player.attack}, 방어력: {player.defense}")

# 몬스터 생성
monsters = [
    Monster("슬라임", 50, 10),
    Monster("고블린", 80, 15),
    Monster("드래곤", 200, 50)
]
current_monster = random.choice(monsters)
current_monster = random.choice(monsters)
print(f"{monsters.name}이(가) 나타났습니다! 체력: {monsters.health}, 공격력: {monsters.attack}")


# 전투 시스템
import random

inventory = ["포션"]

while player.is_alive() and monster.is_alive():
    print(f"\n{player.name}: {player.health} 체력")
    print(f"{monsters.name}: {monsters.health} 체력")

    action = input("공격하러면 '1', 방어하려면 '2', 포션을 사용하려면 '3'을 입력하세요: ")

    if action == "1":
        # 플레이어의 공격
        damage = player.attack
        monsters.take_damage(damage)
        print(f"{player.name}이(가) {monsters.name}에게 {damage} 데미지를 입혔습니다!")
    elif action == "2":
        print(f"{player.name}이(가) 방어 태세를 취합니다!")
    elif action == "3":
        # 포션 사용
        if "포션" in inventory:
            print("꿀꺽! 포션을 사용하여 체력을 20 회복했습니다!")
            player.health += 20
            inventory.remove("포션")    # 포션 사용 후 제거
        else:
            print("포션이 없습니다!")

    # 몬스터의 공격
    if monster.is_alive():
        damage = monster.attack - player.defense
        if damage < 0:
            damage = 0
        player.take_damage(damage)
        print(f"{monster.name}이(가) {player.name}에게 {damage} 데미지를 입혔습니다!")

# 전투 결과
if player.is_alive():
    print(f"\n{player.name}이(가) 승리했습니다!")
else:
    print(f"\n{monsters.name}이(가) 승리했습니다!")

```

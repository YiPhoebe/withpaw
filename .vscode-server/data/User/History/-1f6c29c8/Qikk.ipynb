```python
class Character:
    def __init__(self, name, health, attack, defense, exp):
        self.name = name
        self.health = health
        self.attack = attack
        self.defense = defense
        self.exp = 0
        self.level = 1

    def take_damage(self, damage):
        self.health -= damage
        if self.health < 0:
            self.health = 0

    def is_alive(self):
        return self.health > 0

# 캐릭터 생성
player = Character("유정", 100, 20, 10)
print(f"{player.name} 캐릭터가 생성되었습니다! 체력: {player.health}, 공격력: {player.attack}, 방어력: {player.defense}")

class Monster:
    def __init__(self, name, health, attack):
        self.name = name
        self.health = health
        self.attack = attack

    def take_damage(self, damage):
        self.health -= damage
        if self.health < 0:
            self.health = 0
        
    def is_alive(self):
        return self.health > 0
    
# 몬스터 생성
monster = Monster("슬라임", 50, 10)
print(f"{monster.name}이(가) 나타났습니다! 체력: {monster.health}, 공격력: {monster.attack}")
```

```python
# 전투 시스템

import random

while player.is_alive() and monster.is_alive():
    print(f"\n{player.name}: {player.health} 체력")
    print(f"{monster.name}: {monster.health} 체력")

    action = input("공격하러면 '1', 방어하려면 '2'를 입력하세요: ")

    if action == "1":
        # 플레이어의 공격
        damage = player.attack
        monster.take_damage(damage)
        print(f"{player.name}이(가) {monster.name}에게 {damage} 데미지를 입혔습니다!")
    elif action == "2":
        print(f"{player.name}이(가) 방어 태세를 취합니다. 다음 공격을 막을 준비 완료!")

    # 몬스터의 공격
    if monster.is_alive():
        damage = monster.attack - player.defense
        if damage < 0:
            damage = 0
        player.take_damage(damage)
        print(f"{monster.name}이(가) {player.name}에게 {damage} 데미지를 입혔습니다!")

# 전투 결과
if player.is_alive():
    print(f"\n{player.name}이(가) 승리했습니다!")
else:
    print(f"\n{monster.name}이(가) 승리했습니다!")
```

```python
inventory = ["포션"]
if "포션" in inventory:
    print("포션을 사용하여 체력을 회복합니다!")
    player.health += 20
```

```python
player.exp += 50
if player.exp >= 100:
    player.level += 1
    print(f"{player.name}이(가) 레벨업 했습니다!")
```

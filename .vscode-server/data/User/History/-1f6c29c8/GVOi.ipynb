```python
class Character:
    def __init__(self, name, health, attack, defense):
        self.name = name
        self.health = health
        self.attack = attack
        self.defense = defense

    def take_damage(self, damage):
        self.health -= damage
        if self.health < 0:
            self.health = 0

    def is_alive(self):
        return self.health > 0

# 캐릭터 생성
player = Character("유정", 100, 20, 10))
print(f"{player.name} 캐릭터가 생성되었습니다! 체력: {player.health}, 공격력: {player.attack}, 방어력: {player.defense}")
```

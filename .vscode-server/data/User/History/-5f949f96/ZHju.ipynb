```python
import random

# 캐릭터 클래스
class Character:
    def __init__(self, name, health, attack, defense):
        self.name = name
        self.health = health
        self.attack = attack
        self.defense = defense
        self.exp = 0
        self.level = 1
        self.weapon = "파뿌리" # 캐릭터의 기본 무기

    def take_damage(self, damage):
        self.health -= damage
        if self.health < 0:
            self.health = 0

    def is_alive(self):
        return self.health > 0
    
    def level_up(self):
        self.level += 1
        self.exp = 0
        self.health += 50
        self.attack += 5
        print(f"축하합니다! {self.name}이(가) 레벨 {self.level}로 올라갔습니다!")
        print(f"새로운 체력: {self.health}, 공격력: {self.attack}!")

# 몬스터 클래스
class Monster:
    def __init__(self, name, health, attack):
        self.name = name
        self.health = health
        self.attack = attack

    def take_damage(self, damage):
        self.health -= damage
        if self.health < 0:
            self.health = 0
        
    def is_alive(self):
        return self.health > 0
    
# 캐릭터 생성 (사용자 입력)
print("당신은 누나의 심부름으로 파를 사러 가는 중입니다...")
print("!@#$야~ 딴 길로 새지말고 빨리 다녀와!")
print("누나가 베란다로 당신에게 외치고 있군요.")
print("그런데 당신의 이름이 잘 들리지 않습니다.")
character_name = input("당신의 이름을 입력하세요: ")
player = Character(character_name, 100, 20, 10)
print(f"{player.name} 캐릭터가 생성되었습니다!")
print(f"빠르게 마트에 들러 '파'를 구매했습니다.")
print(f"{player.name}이(가) '파뿌리'를 들고 길을 나섰습니다!")
print(f"체력: {player.health}, 공격력: {player.attack}, 방어력: {player.defense}")

# 몬스터 생성
monsters = [
    Monster("슬라임", 50, 10),
    Monster("고블린", 70, 15),
    Monster("늑대", 80, 20),
    Monster("트롤", 120, 25),
    Monster("드래곤", 200, 50),
]

# 초기 아이템
inventory = ["포션"]

# 게임 루프
while player.is_alive():
    # 랜덤 몬스터 선택
    monster = random.choice(monsters)
    print(f"\n{monster.name}이(가) 나타났습니다! 체력: {monster.health}, 공격력: {monster.attack}")

    while player.is_alive() and monster.is_alive():
        print(f"\n{player.name}: {player.health} 체력")
        print(f"{monster.name}: {monster.health} 체력")

        action = input(
            "매운맛 공격! '1', 뿌리 흙털기! '2', 포션 사용 '3', 도망가기 '4'를 입력하세요: ")
        
        if action == "1":
            # 매운맛 공격: 강한 데미지
            damage = player.attack + 10 # 추가 데미지
            monster.take_damage(damage)
            print(f"{player.name}이(가) '매운맛 공격'으로 {monster.name}에게 {damage} 데미지를 입혔습니다!")
        elif action == "2":
            # 뿌리 흙털기: 약한 데미지 + 몬스터 공격력 감소
            damage = player.attack // 2     # 약한 데미지
            monster.take_damage(damage)
            print(f"{player.name}이(가) '뿌리 흙털기'로 {monster.name}에게 {damage} 데미지를 입혔습니다!")
        elif action == "3":
            # 포션 사용
            if "포션" in inventory:
                print("꿀꺽! 포션을 사용하여 체력을 20 회복했습니다!")
                player.health += 20
                inventory.remove("포션")    # 포션 사용 후 제거
            else:
                print("포션이 없습니다.")
        elif action == "4":
            # 도망가기: 게임 종료
            print(f"{player.name}이(가) {monster.name}으로부터 도망쳤습니다!")
            sys.exit()   # 전투 종료 (도망가기)
        else:
            print("잘못된 입력입니다. 다시 선택하세요.")

        # 몬스터의 공격
        if monster.is_alive():
            # 몬스터가 랜덤한 데미지를 계산
            monster_damage = random.randint(5, monster.attack)  # 랜덤 데미지
            damage = monster_damage - player.defense
            if damage < 0:
                damage = 0
            player.take_damage(damage)
            print(f"{monster.name}이(가) {player.name}에게 {damage} 데미지를 입혔습니다!")
    
    # 전투 결과
    if player.is_alive() and not monster.is_alive():
        print(f"\n{player.name}이(가) {monster.name}을(를) 처치했습니다!")
        player.exp += 50    # 경험치 50 추가
        if player.exp >= 100:
            player.level_up()
    elif not player.is_alive():
        print(f"\n{monster.name}이(가) {player.name}을(를) 이겼습니다!")
        break   # 플레이어 사망 시 게임 종료

print("\n게임종료! 용감히 싸운 당신을 응원합니다!")
```

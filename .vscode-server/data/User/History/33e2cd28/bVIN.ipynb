```python
# 첫 번째 숫자를 입력하세요: 5
# 두 번째 숫자를 입력하세요: 3

number1 = int(input("첫 번째 숫자를 입력하세요: "))
number2 = int(input("두 번째 숫자를 입력하세요: "))
num = number1 + number2
print(f"두 숫자의 합은: {num}입니다.")
```

```python
num = int(input("숫자를 입력하세요: "))
result = sum(range(1, num + 1))
print(f"1부터 {num}까지의 합은 {result}입니다.")
```

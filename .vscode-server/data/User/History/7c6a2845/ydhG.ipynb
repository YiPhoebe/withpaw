```python
import os
from pathlib import Path

file_path = "example.txt"

# os.stat() 사용
file_stat = os.stat(file_path)
print("파일 크기:", file_stat.st_size)  # 파일 크기 출력
print("마지막 수정 시간:", file_stat.st_mtime)  # 마지막 수정 시간

# pathlib 사용
path = Path(file_path)
print("파일 크기:", path.stat().st_size)
print("마지막 접근 시간:", path.stat().st_atime)

# 파일을 읽기 전용으로 설정
os.chmod(file_path, 0o444)  # 읽기 전용 권한 설정

# 파일에 쓰기 권한이 있는지 확인
if os.access(file_path, os.W_OK):
    print("파일에 쓰기 권한이 있습니다.")
else:
    print("파일에 쓰기 권한이 없습니다.")
```

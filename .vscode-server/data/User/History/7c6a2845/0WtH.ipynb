```python
try:
    1 / 0
except ZeroDivisionError:
    print("예외 정보 : ", sys.exc_info())
```

```python
try:
    result = 10 / 0     # ZeroDivisionError
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
else:
    print("예외가 발생하지 않았습니다.")
finally:
    print("예외 처리 완료")
```

```python
try:
    raise ValueError("잘못된 값")
except ValueError as e:
    raise TypeError("타입 오류")
```

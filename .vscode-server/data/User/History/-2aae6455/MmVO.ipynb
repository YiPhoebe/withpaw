```python
import random
import sys

# 캐릭터 클래스
class Character:
    def __init__(self, name, health, attack, defense):
        self.name = name
        self.health = health
        self.attack = attack
        self.defense = defense
        self.exp = 0
        self.level = 1
        self.weapon = "파뿌리"  # 캐릭터 기본 무기
    
    def take_damage(self, damage):
        self.health -= damage
        if self.health < 0:
            self.health = 0

    def is_alive(self):
        return self.health > 0
    
    def level_up(self):
        self.level += 1
        self.exp = 0
        self.health += 50
        self.attack += 5
        print(f"축하합니다! {self.name}이(가) 레벨 {self.level}로 올라갔습니다!")
        print(f"새로운 체력: {self.health}, 공격력: {self.attack}!")

# 몬스터 클래스
class Monster:
    def __init__(self, name, health, attack):
        self.name = name
        self.initial_health = health
        self.health = health
        self.attack = attack

    def take_damage(self, damage):
        self.health -= damage
        if self.health < 0:
            self.health = 0

    def is_allive(self):
        return self.health > 0
    
    def reset_health(self):
        self.health = self.initial_health   # 초기 체력으로 복구

# 스토리와 단계 설정
story_stages = [
    {"monster": Monster("슬라임", 50, 10), "description": "슬라임이 길을 막고 있습니다."},
    {"monster": Monster("고블린", 70, 15), "description": "고블린이 당신을 가로막습니다."},
    {"monster": Monster("늑대", 80, 20), "description": "늑대가 사나운 눈빛으로 다가옵니다."},
    {"monster": Monster("트롤", 120, 25), "description": "거대한 트롤이 길을 지키고 있습니다."},
    {"monster": Monster("드래곤", 200, 50), "description": "최종 보스 드래곤이 하늘에서 내려옵니다!"},
]

# 초기 아이템
inventory = ["포션"]

# 캐릭터 생성 (사용자 입력)
print("당신은 누나의 심부름으로 파를 사러 가는 중입니다...")
print("!@#$야~ 딴 길로 새지말고 빨리 다녀와!")
print("누나가 베란다로 당신에게 외치고 있군요.")
print("그런데 당신의 이름이 잘 들리지 않습니다.")
character_name = input("당신의 이름을 입력하세요: ")
player = Character(character_name, 100, 20, 10)
print(f"{player.name} 캐릭터가 생성되었습니다!")
print(f"빠르게 마트에 들러 '파'를 구매했습니다.")
print(f"{player.name}이(가) '파뿌리'를 들고 길을 나섰습니다!")
print(f"체력: {player.health}, 공격력: {player.attack}, 방어력: {player.defense}")

# 게임 루프 : 스토리 단계 진행
for stage, stage_info in enumerate (story_stages, start=1):
    monster = stage_info["monster"]
    monster.reset_health()   # 몬스터 체력 초기화
    print(f"\n[스테이지 {stage}] {stage_info['description']}")
    print(f"{monster.name}이(가) 나타났습니다! 체력: {monster.health}, 공격력: {monster.attack}")

    # 전투 시스템
    while player.is_alive() and monster.is_alive():
        print(f"\n{player.name}: {player.health} 체력")
        print(f"{monster.name}: {monster.health} 체력")

        action = input(
            "매운맛 공격! '1', 뿌리 흙털기! '2', 포션 사용 '3', 도망가기 '4'를 입력하세요: ")
        
        if action == "1":
            # 매운맛 공격: 강한 데미지
            damage = player.attack + 10 # 추가 데미지
            monster.take_damage(damage)
            print(f"{player.name}이(가) '매운맛 공격!'으로 {monster.name}에게 {damage} 데미지를 입혔습니다.")
        elif action == "2":
            # 뿌리 흙털기: 약한 데미지 + 몬스터 공격력 감소
            damage = player.attack // 2     # 약한 데미지
            monster.take_damage(damage)
            print(f"{player.name}이(가) '뿌리 흙털기!'로 {monster.name}에게 {damage} 데미지를 입혔습니다.")
        elif action == "3":
            # 포션 사용
            if "포션" in inventory:
                print("꿀꺽! 포션을 사용하여 체력을 20 회복했습니다!")
                player.health += 20
                inventory.remove("포션")    # 포션 사용 후 제거
            else:
                print("포션이 없습니다.")
        elif action == "4":
            # 도망가기: 게임 종료
            print(f"{player.name}이(가) {monster.name}으로부터 도망쳤습니다.")
            print("하지만 몬스터가 파를 빼앗아 버렸습니다. 게임 종료!")
            sys.exit()      # 전투 종료 (도망가기)
        else:
            print("잘못된 입력입니다. 다시 선택하세요.")

        # 몬스터의 공격
        if monster.is_alive():
            # 몬스터가 랜덤한 데미지를 계산
            monster_damage = random.randint(5, monster.attack)  # 랜덤 데미지
            damage = monster_damage - player.defense
            if damage < 0:
                damage = 0
            player.take_damage(damage)
            print(f"{monster.name}이(가) {player.name}에게 {damage} 데미지를 입혔습니다.")

    # 전투 결과
    if player.is_alive() and not monster.is_alive():
        print(f"\n{player.name}이(가) {monster.name}을(를) 처치했습니다!")
        player.exp += 50
        if player.exp >= 100:
            player.level_up()
        # 포션 드롭: 50% 확률
        if random.random() < 0.5:
            print(f"\n{monster.name}이(가) 포션을 떨어뜨렸습니다!")
            inventory.append("포션")
    elif not player.is_alive():
        print(f"\n{monster.name}이(가) {player.name}을(를) 쓰러뜨렸습니다.")
        print("\n몬스터가 파를 빼앗아 가 버렸습니다...")
        print("\n누나: 이게 뭐여? 파도 안 사고오고 뭐했노 이복동!!")
        print("\n게임 종료! 더 강해져서 도전하세요!")
        sys.exit()  # 플레이어 사망 시 게임 종료

# 스토리 완료
print("\n축하합니다! 모든 몬스터를 처치하고 누나에게 파를 전달했습니다.")
print("누나 : 이제야 들어왔나?? 빨리 들어온나~!")
```
